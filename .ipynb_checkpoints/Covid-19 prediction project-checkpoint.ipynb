```python
import pandas as pd
import numpy as np
import time
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
```

## 1. Data cleaning and merging

```python
## Import datasets
data_2020_mob = pd.read_csv("2020_US_Region_Mobility_Report.csv")
data_2021_mob = pd.read_csv("2021_US_Region_Mobility_Report.csv")
data_2022_mob = pd.read_csv("2022_US_Region_Mobility_Report.csv")
data_covid = pd.read_csv("United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv")
data_population = pd.read_csv("population density.csv")
data_vulnerable = pd.read_excel("vulnerability.xlsx", sheet_name="ccvi-US-state")
```

```python
## Clean vulnerability data
data_vulnerable.rename({'stateName':'state',
                       'ccvi':'vulnerability',
                       'theme1':'socioeconomic',
                       'theme2':'minority',
                       'theme3':'household',
                       'theme4':'epidemiological',
                       'theme5':'healthcare',
                       'theme6':'environment',
                       'theme7':'density_v'}, axis=1, inplace=True)
data_vulnerable['state'] = data_vulnerable.state.str.title()
```

```python
## Clean population data
data_population = data_population.iloc[2:, 1:4]
data_population.columns = ['state', 'population', 'density']
```

```python
## Clean and view covid data: we have one entry per state per day
data_covid['date'] = pd.to_datetime(data_covid.submission_date)
data_covid.groupby('state').apply(pd.DataFrame.sort_values, 'date') 
```

```python
## Clean mobility data

data_2020_mob = data_2020_mob[(data_2020_mob['sub_region_1'].notnull()) & (data_2020_mob['sub_region_2'].isnull())]
data_2021_mob = data_2021_mob[(data_2021_mob['sub_region_1'].notnull()) & (data_2021_mob['sub_region_2'].isnull())]
data_2022_mob = data_2022_mob[(data_2022_mob['sub_region_1'].notnull()) & (data_2022_mob['sub_region_2'].isnull())]

def clean_mob_data(df):
    df['state'] = df['iso_3166_2_code'].str.slice(3)
    df['date'] = pd.to_datetime(df.date)
    df = df.rename(columns = {'sub_region_1':'state full'})
    df.columns = df.columns.str.replace('_percent_change_from_baseline', '')
    return df.drop(['country_region_code', 'country_region', 'sub_region_2', 'metro_area', 'census_fips_code', 'place_id', 'iso_3166_2_code'], axis = 1)

mob_data_list = [data_2020_mob, data_2021_mob, data_2022_mob]
mob_data_list = [clean_mob_data(df) for df in mob_data_list]
```

```python
# Merge covid + population data with mobility data

merged_data_list = [df.merge(data_covid[['state', 'date', 'tot_cases', 'new_case']], 
                             how = 'left', 
                             on = ['state', 'date']) for df in mob_data_list]
merged_data_list1 = [df.merge(data_vulnerable.iloc[:, 1:], 
                             how = 'left',
                             left_on = 'state full',
                             right_on = 'state') for df in merged_data_list]
merged_data_list2 = [df.merge(data_population, 
                             how = 'left',
                             left_on = 'state full',
                             right_on = 'state') for df in merged_data_list1]
```

## 2.  Data Visualizations

### Pearson coefficients

```python
## Global variable for window size
days = 15
```

```python
## Get Pearson coefficient between previous X days mobility and next X days cases for each row in the data
def get_pearson(df, category):
    df.reset_index(drop=True, inplace=True)
    for i in range(days, len(df.index)-days):
        next_cases = df.loc[i:i + days, 'new_case'].values
        past_mobility = df.loc[i - days:i, category].values
        df.loc[i, 'pearson_' + category] = np.corrcoef(next_cases, past_mobility)[0][1]
        
    return df
    
def get_p_bystate(df):
    df_return = None
    for state in df.state_y.unique():
        df_state = df[df['state_y'] == state].sort_values(by = 'date')
        for category in ['retail_and_recreation', 'grocery_and_pharmacy', 'parks', 'transit_stations', 'workplaces', 'residential']:
            df_state = get_pearson(df_state, category)

        if df_return is None:
            df_return = df_state
        else:
            df_return = pd.concat([df_return, df_state], ignore_index=True)
    
    return df_return
    
pearson_data_list = [get_p_bystate(df) for df in merged_data_list2]
```

```python
data_all = pd.concat([pearson_data_list[0], pearson_data_list[1], pearson_data_list[2]], ignore_index=True)
```

### Density and infection rate

```python
data_all2 = data_all.copy()
data_all2.replace(',', '', regex=True, inplace=True)
data_all2['density'] = data_all2['density'].astype(float)

conditions = [
    (data_all2['density'] < 70),
    (data_all2['density'] >= 70) & (data_all2['density'] < 150),
    (data_all2['density'] > 150)
]
values = ['low', 'medium', 'high']
data_all2['density_category'] = np.select(conditions, values)
```

```python
def plot_infection_by_density(mobility_category):
    plot_data = data_all2.groupby(['density_category', mobility_category]).mean()['new_case'].reset_index()

    sns.set(rc={'figure.figsize':(16,9)})
    sns.lineplot(data=plot_data, x=mobility_category, y='new_case', hue='density_category')

plot_infection_by_density('transit_stations')
```

## Prepare data to build models

```python
## Rate of new cases (instead of abs number), not used but in case we need later
data_all2['new_case_rate'] = data_all2['new_case']/data_all2['population'].astype(int)*100

## Generate our y labels: label each row as either small/large increase or small/large decrease
def generate_label(df):
    df.reset_index(drop=True, inplace=True)
    for i in range(days, len(df.index)-days):
        next_cases = np.mean(df.loc[i:i + 14, 'new_case'].values) ## We are always predicting change in next 14 days
        past_cases = np.mean(df.loc[i - days:i, 'new_case'].values)
        diff = next_cases - past_cases
        if diff > 1000:
            df.loc[i, 'label'] = 'large increase'
        elif (diff > 0 and diff <= 1000):
            df.loc[i, 'label'] = 'small increase'
        elif (diff < 0 and diff >= -1000):
            df.loc[i, 'label'] = 'small decrease'
        elif (diff < -1000):
            df.loc[i, 'label'] = 'large decrease'
    
    return df
    
## Add columns for previous X days mobility and new cases. 
def get_label_bystate(df):
    df_return = None
    for state in df.state_y.unique():
        df_state = df[df['state_y'] == state].sort_values(by = 'date')
        generate_label(df_state)
        
        ##add columns for previous X days mobility
        mobility_categories = ['retail_and_recreation', 'grocery_and_pharmacy', 'parks', 'transit_stations', 
                               'workplaces', 'residential']
        for category in mobility_categories:
            for i in range(days):
                df_state[category + '_' + str(i) + '_'] = df_state[category].shift(periods=i)
                
        ## Add columns for previous X days new cases
        for j in range(days):
            df_state['new_case' + '_' + str(j) + '_'] = df_state['new_case'].shift(periods=j)

        if df_return is None:
            df_return = df_state
        else:
            df_return = pd.concat([df_return, df_state], ignore_index=True)
    
    return df_return
    
data_all3 = get_label_bystate(data_all2)
data_all3.dropna(inplace=True)
data_all3.drop(['retail_and_recreation', 'grocery_and_pharmacy', 'parks', 'transit_stations', 
                               'workplaces', 'residential', 'new_case'], axis = 1, inplace=True)
```

### SHAP values

Def: For each data point, take each feature and calculate its marginal contribution to the predicted value across all feature combinations.

Multicolinearity does NOT affect the performance for Random Forest models. However, due to the random feature selection in the model, it may rely on one correlated variable rather than another for no reason. Since SHAP interprets feature importance of the model (not the data), this means SHAP may still assign different scores to highly correlated features A and B, depending on the underlying seed of the model.

### Permutation importance
Definition: For each tree, get test accuracy with all features, then randomly permute feature i and compute the drop in accuracy. Average this drop across all trees to get the importance of the feature i.
- Results can be very misleading if there are features with high correlation. e.g. if A and B are highly correlated and A is randomly permuted, model can just use B and perform nearly as well. Hence, score for highly correlated features are all deceptively close to 0.

```python
from sklearn.inspection import permutation_importance
```

```python
X_permute = X_test.sample(300)
y_permute = y_test.loc[X_permute.index]
result = permutation_importance(forest, X_permute, y_permute, n_repeats=10, random_state=42)
perm_sorted_idx = result.importances_mean.argsort()

fig, ax = plt.subplots(figsize=(15, 8))
ax.boxplot(
    result.importances[perm_sorted_idx].T,
    labels=X_permute.columns[perm_sorted_idx],
)
plt.xticks(rotation='vertical')
plt.show()
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV

extra_forest.get_params()
```

```python
## number of trees
n_estimators = [int(x) for x in np.linspace(100, 1000, 10)]
## number of features considered per split (lower = greater reduction in variance)
max_features = ['sqrt', 'log2', 0.3, None] ## "more randomness can be achived by max_features < 1, 0.3 is typical value in literature"
## max depth per tree
max_depth = [int(x) for x in np.linspace(10, 100, 10)]
max_depth.append(None)
# min samples required to split a node
min_samples_split = [2, 5, 10]
## boostrap
bootstrap = [True, False]

# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'bootstrap': bootstrap}

print(random_grid)
```

### RNN

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import models
from tensorflow.keras.optimizers import SGD
```

```python
## Columns that are present in every data point at every time step
perm_columns = ['vulnerability', 'socioeconomic', 'minority', 'household',
       'epidemiological', 'healthcare', 'environment', 'density_v',
       'population', 'density', 'pearson_retail_and_recreation',
       'pearson_grocery_and_pharmacy', 'pearson_parks',
       'pearson_transit_stations', 'pearson_workplaces', 'pearson_residential']

## Convert X_train/X_test data into rnn format
## e.g. Let f1_d represent feature 1 on day d. Each row in X_train will correspond to a nested list
## [[f1_1, f2_1...], [f1_2, f2_2...], ..., [f1_15, f2_15, ..., f23_15]]
def get_x_rnn(X):
    X_rnn = []
    for i in X.index: ## For every row of data in X...
        curr_row = X.loc[i, :]
        data_i = [] ## Build a list with 15 timesteps, each timestep is represented with a list of 23 features
        for j in range(days):
            step_j_features = perm_columns + [col for col in X.columns if '_' + str(j) + '_' in col]
            step_j_data = np.array(curr_row[step_j_features])
            data_i.append(step_j_data)
        X_rnn.append(data_i) 
    return X_rnn
    
X_train_rnn = get_x_rnn(X_train)
X_test_rnn = get_x_rnn(X_test)
## Convert to numpy array format for keras
X_train_rnn = np.asarray([np.asarray(sublist) for sublist in X_train_rnn]).astype(np.float32)
X_test_rnn = np.asarray([np.asarray(sublist) for sublist in X_test_rnn]).astype(np.float32)

## Convert y_train/y_test into rnn format. Each label corresponds to an array of size 1. 
def convert_y_to_number(label):
    if label == 'small increase':
        return 2
    elif label == 'large increase':
        return 3
    elif label == 'small decrease':
        return 1
    elif label == 'large decrease':
        return 0
    else:
        return 'LABEL ERROR'

y_train_rnn = np.asarray([np.array(convert_y_to_number(elem)) for elem in y_train]).astype(np.float32)
y_test_rnn = np.asarray([np.array(convert_y_to_number(elem)) for elem in y_test]).astype(np.float32)
```

```python
## Another attempt to incorporate the correct loss function, based on: https://machinelearningmastery.com/how-to-choose-loss-functions-when-training-deep-learning-neural-networks/
model = keras.models.Sequential()
model.add(layers.Input(shape=(15, 23)))
model.add(tf.keras.layers.Flatten()) 
model.add(layers.Dense(50, activation='relu', kernel_initializer='he_uniform'))
model.add(layers.Dense(4, activation='softmax'))

opt = SGD(lr=0.01, momentum=0.9)
model.compile(loss='sparse_categorical_crossentropy', optimizer=opt, metrics=['accuracy'])

model.fit(X_train_rnn, y_train_rnn)
```
